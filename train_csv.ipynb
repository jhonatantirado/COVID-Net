```python
import numpy as np
import pandas as pd
import os
from shutil import copyfile
```

```python
train_file = 'train_split_v3.csv'
```

```python
data = pd.read_csv(train_file)
data.head(5)
```

```python
idx_covid19 = data["label"] == "COVID-19"
covid19_data = data [idx_covid19]
print(len(covid19_data.index))
covid19_data.head(5)
```

```python
idx_pneumonia = data["label"] == "pneumonia"
pneumonia_data = data [idx_pneumonia]
print(len(pneumonia_data.index))
pneumonia_data.head(5)
```

```python
idx_normal = data["label"] == "normal"
normal_data = data [idx_normal]
print(len(normal_data.index))
normal_data.head(5)
```

```python
final = pd.concat([covid19_data,pneumonia_data[0:93],normal_data[0:93]])
print(len(final.index))
file_name = 'final.csv'
final.to_csv(file_name, sep=',', encoding='utf-8',index=False)
```

```python
origin_path = 'data/train'
destination_path = 'M1_xray_c19'

for index, row in final.iterrows():
    print(row['file'], row['label'])
    origin = os.path.join(origin_path, row['file'])
    destination = os.path.join(destination_path, row['label'], row['file'])
    print (origin)
    print (destination)
    copyfile(origin, destination)
```
